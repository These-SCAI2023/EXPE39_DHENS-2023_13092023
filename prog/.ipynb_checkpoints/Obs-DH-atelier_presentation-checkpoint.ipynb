# Named Entity Recognition (NER) on noisy data with spaCy

https://spacy.io/

Spacy is an automatic language processing tool that can also be used for Named Entity Recognition (NER).

It can be used to retrieve Named Entities accompanied by a Label, as well as the position of the start and end character of the identified term. 

Place name : LOC ; 
Person name : PERS ; 
Organisations : ORG ; ...

```python
import spacy
import time
from spacy import displacy
import glob

```

```python
def read_text(path_corpora):
        with open (path, "r", encoding="utf-8") as fichier:
            txt=fichier.read()
        return txt[:2000]
path_corpus= "../DATA/ELTeCFRA/*/*.txt"
# Load the language model you are working with
modele="fr_core_news_lg"
nlp = spacy.load(modele)
```

## Retrieve NER output in tabular form with pandas 



```python
import pandas as pd
from IPython.display import display
import json
```

```python
def read_text(path_corpora):
        with open (path, "r", encoding="utf-8") as fichier:
            txt=fichier.read()
        return txt
    
def stocker(chemin, contenu):
    w =open(chemin, "w")
    w.write(json.dumps(contenu , indent = 2))
    w.close()
    # print(chemin)
    
    return chemin
```

```python
for path in glob.glob(path_corpus):

    dict_ent={}
    liste_ent=[]
    liste_label=[]
    liste_contxt_droit=[]
    liste_contxt_gauche=[]
    version=path.split("/")[3]
    auteur=version.split("_")[0]
    version=version.split("_")[-1]
    print("------------",auteur,version)

    
    text=read_text(path)
    #print(f"          \n\n {text}")
    
    doc = nlp(text)

    for ent in doc.ents:
        if ent.label_=="LOC":### Retrieve spacial entities only
            liste_ent.append(ent.text)
            liste_label.append(ent.label_)
            liste_contxt_gauche.append(text[ent.start_char-10:ent.start_char])

            liste_contxt_droit.append(text[ent.end_char:ent.end_char+10])
        
    dict_ent["Entite"]=liste_ent
    dict_ent["Label"]=liste_label
    dict_ent["liste_contxt_gauche"]=liste_contxt_gauche
    dict_ent["liste_contxt_droit"]=liste_contxt_droit
    
    data_tab = pd.DataFrame(dict_ent)
    
    display(data_tab)  
    data_tab.to_csv(f"../DATA/ELTeCfra{version}.csv", sep=";")
    stocker("%s-%s.json"%(path,modele),liste_ent) 
        
       
    
```

# Evaluation of NER output on noisy data : intersections



```python
import re
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles

def read_json (chemin):
    with open(chemin) as json_data: 
        dist =json.load(json_data)
        return dist
    
def diagramme_venn(liste_en_pp, liste_en_ocr,ocr):
    font2 = {'size': 25} # use for labels
    plt.rc('font', **font2) # sets the default font 
    plt.rcParams['text.color'] = 'black' # changes default text colour
    venn2([set(liste_en_pp), set(liste_en_ocr)],set_labels = ('EN Réf', 'EN %s'%ocr),set_colors=("darkgrey", "darkblue"),alpha=0.5)

    venn2_circles(subsets=(set(liste_en_pp), set(liste_en_ocr)),linestyle="dotted", linewidth=1) 
    intersection=set(liste_en_pp)&set(liste_en_ocr)
    print(ocr)
    print(intersection)
    
    
     
def stocker(nom_fichier):
    plt.gcf().set_size_inches(12, 7)
    plt.subplots_adjust(left=0.03, right=0.8, top=0.9, bottom=0.1)
    plt.savefig(nom_fichier, dpi=300)
    plt.clf()
    return nom_fichier
 
    
path_data= "../DATA/ELTeCFRA/*/*.json"
dict_global={}
liste_ocr=[]
liste_entit=[]
liste_vers=[]

for path in glob.glob(path_data):
    data=read_json(path)
    
    vers=path.split("/")[3]
    vers=vers.split("_")[-1]
    
    if vers!="REF":
        liste_ocr.append(vers)
        
    

    for entity in data:
            liste_entit.append(entity)
            liste_vers.append(vers)
            
    
        
dict_global["Entite"]=liste_entit
dict_global["version"]=liste_vers
data_tab = pd.DataFrame(dict_global)
  
for name_ocr in liste_ocr:
        
    
 
    data_ref=data_tab.query("version=='REF'" )
    data_ocr=data_tab.query("version== @name_ocr" )
   
    
    diagramme_venn(data_ref.Entite,data_ocr.Entite,name_ocr)
    plt.show()
    stocker("../DATA/INTERSECTION_GLOBALES/%s_%s_intersection.png"%(name_ocr,modele))  
   
    


```

```python
data_ocr1=data_tab.query("version==@liste_ocr[1]" )
data_ocr2=data_tab.query("version== @liste_ocr[0]" )


diagramme_venn(data_ocr1.Entite,data_ocr2.Entite,name_ocr)
plt.show()
stocker("../DATA/INTERSECTION_GLOBALES/%s_%s_intersection2.png"%(name_ocr,modele)) 
```

```python
import re
import sklearn
from sklearn.neighbors import DistanceMetric
from sklearn.feature_extraction.text import CountVectorizer

def stocker( chemin, contenu):

    w =open(chemin, "w")
    w.write(json.dumps(contenu , indent = 2))
    w.close()
    print(chemin)
    

def get_distances_word(texte1, texte2, N=1, liste_name =["jaccard", "braycurtis","dice", "cosinus"] ):
    dico = {}
    for metric_name in liste_name :
        dico[metric_name] = []
        liste_resultat_dist2 = []
#        for n_max in range(1, N+1):###range([min, default = 0], max, [step, default = 1]) 
        V = CountVectorizer(analyzer='word')
        X = V.fit_transform([texte1, texte2]).toarray()
        if metric_name!= "cosinus" :  
            dist = DistanceMetric.get_metric(metric_name)     
            distance_tab1=dist.pairwise(X)
            liste_resultat_dist2.append(distance_tab1[0][1])
        else: 
            distance_tab1=sklearn.metrics.pairwise.cosine_distances(X) 
            liste_resultat_dist2.append(distance_tab1[0][1])
        dico[metric_name] = liste_resultat_dist2
    return dico

## MAIN

path_corpus="../DATA/ELTeCFRA/*/*"

for subcorpus in glob.glob(path_corpus):
    # print(subcorpus)
    
    outputs = glob.glob(f"{subcorpus}/*/*")
    dico_out = {}
    dist_txt = {}
    for file_type in ["txt", ".json"]:
        chemins = [x for x in outputs if len(re.findall(f"{file_type}$",x))>0]
        print(chemins)
        liste_compare = []
        for path_file in chemins:
            filename = re.split("/", path_file)[-1]
            elems = re.split("_|\\.", filename)
            print("ELEMS",elems)
            
            auteur,titre, version,correction, modele = elems[0], elems[1],elems[2], elems[3],elems[-2]
            # print(version, correction, modele)
            if file_type =="txt":
                liste_compare.append([version+"-"+correction, read_text(path_file)])
                
            else:
                with open (path_file) as f :
                    liste = json.load(f)
                chaine = " ".join(liste)
                liste_compare.append([ modele, chaine])
               
        for c in liste_compare:
            print("file_type, c[0]",file_type, c[0])
        dico_out[file_type] = {}
        for ID1 in range(len(liste_compare)):
            # print("ID1",ID1)
            version1 = liste_compare[ID1][0]
            # print("version1",version1)
            for ID2 in range(ID1+1, len(liste_compare)):
                # print("ID2",ID2)
                version2 = liste_compare[ID2][0]
                print("version2",version2)
                dico_dist = get_distances_word(liste_compare[ID1][-1], liste_compare[ID2][-1], N=5)
                paire = "%s--%s"%(version1, version2)
                dico_out[file_type][paire] = dico_dist
    
    
       
    stocker("%s_dist-word.json"%(subcorpus), dico_out)
#  
       
        
    
  
```
